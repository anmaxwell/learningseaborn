```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Read the file into a variable fifa_data
fifa_data = pd.read_csv('fifa.csv', index_col="Date", parse_dates=True)
fifa_data.tail()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(24,6))

# Line chart showing how FIFA rankings evolved over time 
sns.lineplot(data=fifa_data)
```

```python
fifa_data.columns
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Fifa positions")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=fifa_data['BRA'], label="Brazil")

# Line chart showing daily global streams of 'Despacito'
sns.lineplot(data=fifa_data['ESP'], label="Spain")

# Add label for horizontal axis
plt.xlabel("Year")
# Add label for vertical axis
plt.ylabel("Position")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))
# Add title
plt.title("Brazil Positions")
# Bar chart showing each position for Brazil
sns.barplot(x=fifa_data.index, y=fifa_data['BRA'])
# Add label for vertical axis
plt.ylabel("Position")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))
# Add title
plt.title("Positions")
# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=fifa_data, annot=False)
# Add label for horizontal axis
plt.xlabel("Country")
# Add label for vertical axis
plt.ylabel("Date")
```

```python
# Read in new data
ins_data = pd.read_csv('insurance.csv')
ins_data.tail()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))
# Add title
plt.title("Scatter Plot")
# Scatterplot showing bmi and charges
sns.scatterplot(x=ins_data['bmi'], y=ins_data['charges'])
# Add label for horizontal axis
plt.xlabel("BMI")
# Add label for horizontal axis
plt.ylabel("Charges")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))
# Add title
plt.title("Scatter Plot")
# Scatterplot showing bmi and charges
sns.scatterplot(x=ins_data['bmi'], y=ins_data['charges'], hue=ins_data['smoker'])
sns.lmplot(x="bmi", y="charges", hue="smoker", data=ins_data)
# Add label for horizontal axis
plt.xlabel("BMI")
# Add label for horizontal axis
plt.ylabel("Charges")
```

```python
sns.swarmplot(x=ins_data['smoker'],
              y=ins_data['charges'])
```
